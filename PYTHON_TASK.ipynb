Write a Python program that takes a sentence and:
Counts vowels and consonants and reverses the sentence and converts it to camel case


```
import re

def process_sentence(sentence):
    """
    Processes a sentence to count vowels and consonants, reverse it, and convert to camel case.

    Args:
        sentence: The input sentence string.

    Returns:
        A tuple containing:
        - vowel_count: The number of vowels in the sentence.
        - consonant_count: The number of consonants in the sentence.
        - reversed_sentence: The sentence reversed.
        - camel_case_sentence: The sentence converted to camel case.
    """
    vowels = "aeiouAEIOU"
    vowel_count = sum(1 for char in sentence if char in vowels)
    consonant_count = sum(1 for char in sentence if char.isalpha() and char not in vowels)

    reversed_sentence = sentence[::-1]

    # Convert to camel case
    words = re.findall(r'\b\w+\b', sentence.lower())
    camel_case_sentence = "".join(word.capitalize() for word in words)
    if camel_case_sentence:
        camel_case_sentence = camel_case_sentence[0].lower() + camel_case_sentence[1:]

    return vowel_count, consonant_count, reversed_sentence, camel_case_sentence

# Example usage:
sentence = "This is a sample sentence."
vowel_count, consonant_count, reversed_sentence, camel_case_sentence = process_sentence(sentence)

print(f"Original sentence: {sentence}")
print(f"Vowel count: {vowel_count}")
print(f"Consonant count: {consonant_count}")
print(f"Reversed sentence: {reversed_sentence}")
print(f"Camel case sentence: {camel_case_sentence}")
```

 Create a tuple of 10 random numbers. Then:
Find the max and min
and Count how many times each number appears and Convert it to a list and sort it


```
import random
from collections import Counter

# Create a tuple of 10 random numbers
random_numbers_tuple = tuple(random.randint(1, 100) for _ in range(10))
print(f"Original tuple: {random_numbers_tuple}")

# Find the max and min
max_number = max(random_numbers_tuple)
min_number = min(random_numbers_tuple)
print(f"Maximum number: {max_number}")
print(f"Minimum number: {min_number}")

# Count how many times each number appears
number_counts = Counter(random_numbers_tuple)
print(f"Number counts: {number_counts}")

# Convert it to a list and sort it
random_numbers_list = list(random_numbers_tuple)
random_numbers_list.sort()
print(f"Sorted list: {random_numbers_list}")
```

 Prompt:Create a class CyberThreatDetector with:
Attributes: source, threat_level
and Method to classify threat as low/medium/high
and Method to display threat summary
in this way display oops concept example

```
class CyberThreatDetector:
    """
    A class to represent a cyber threat detector and classify threat levels.
    """
    def __init__(self, source, threat_level_value):
        """
        Initializes the CyberThreatDetector object.

        Args:
            source (str): The source of the cyber threat.
            threat_level_value (int): A numerical value representing the threat level.
        """
        self.source = source
        self.threat_level_value = threat_level_value
        self.threat_level = self.classify_threat()

    def classify_threat(self):
        """
        Classifies the threat level as low, medium, or high based on the numerical value.

        Returns:
            str: The classified threat level ('Low', 'Medium', or 'High').
        """
        if self.threat_level_value < 40:
            return "Low"
        elif 40 <= self.threat_level_value < 70:
            return "Medium"
        else:
            return "High"

    def display_threat_summary(self):
        """
        Displays a summary of the cyber threat.
        """
        print(f"Threat Source: {self.source}")
        print(f"Threat Level: {self.threat_level}")
```

```
# Create an instance of the class
threat1 = CyberThreatDetector("Network Scan", 55)

# Display the threat summary
threat1.display_threat_summary()

print("-" * 20)

# Create another instance
threat2 = CyberThreatDetector("Phishing Email", 85)

# Display the threat summary for the second instance
threat2.display_threat_summary()
```
